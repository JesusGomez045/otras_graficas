```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para el gráfico
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']
valores = [15, 30, 45, 60]

# Crear el gráfico de barras horizontales
plt.barh(categorias, valores)

# Etiquetas y título
plt.xlabel('Valores')
plt.ylabel('Categorías')
plt.title('Gráfico de Barras Horizontales')

# Mostrar el gráfico
plt.show()



# Datos para el gráfico
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']
valores_set1 = [15, 30, 45, 60]
valores_set2 = [10, 25, 35, 55]

# Definir el ancho de las barras
ancho_barra = 0.35

# Coordenadas para las barras
x = np.arange(len(categorias))

# Crear el gráfico de barras dobles
plt.bar(x - ancho_barra/2, valores_set1, width=ancho_barra, label='Conjunto 1')
plt.bar(x + ancho_barra/2, valores_set2, width=ancho_barra, label='Conjunto 2')

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras Dobles')
plt.xticks(x, categorias)
plt.legend()

# Mostrar el gráfico
plt.show()

# Datos para el gráfico
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# Crear el gráfico de líneas
plt.plot(x, y, marker='o', linestyle='-')

# Etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas')

# Mostrar el gráfico
plt.grid(True)  # Puedes quitar esto si no deseas líneas de cuadrícula
plt.show()

# Definir la función lineal
m = 2  # Pendiente
b = 3  # Ordenada al origen

# Generar valores de x
x = np.linspace(0, 10, 100)  # Valores de x de 0 a 10

# Calcular los valores correspondientes de y
y = m * x + b

# Crear el gráfico de la función
plt.plot(x, y, label=f'y = {m}x + {b}')

# Etiquetas y título
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de una Función Lineal')
plt.legend()

# Mostrar el gráfico
plt.grid(True)  # Puedes quitar esto si no deseas líneas de cuadrícula
plt.show()

# Datos para el gráfico de pastel
categorias = ['Categoría A', 'Categoría B', 'Categoría C', 'Categoría D']
valores = [25, 30, 15, 10]

# Colores para las porciones del pastel
colores = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Destacar una porción (opcional)
explode = (0.1, 0, 0, 0)

# Crear el gráfico de pastel
plt.pie(valores, labels=categorias, colors=colores, explode=explode, autopct='%1.1f%%', startangle=90)

# Título
plt.title('Gráfico de Pastel')

# Mostrar el gráfico
plt.axis('equal')  # Aspecto igual para que el gráfico sea un círculo
plt.show()
```
