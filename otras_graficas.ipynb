```python
import matplotlib.pyplot as plt
import numpy as np

# Datos para el grfico
categorias = ['Categora A', 'Categora B', 'Categora C', 'Categora D']
valores = [15, 30, 45, 60]

# Crear el grfico de barras horizontales
plt.barh(categorias, valores)

# Etiquetas y ttulo
plt.xlabel('Valores')
plt.ylabel('Categoras')
plt.title('Grfico de Barras Horizontales')

# Mostrar el grfico
plt.show()



# Datos para el grfico
categorias = ['Categora A', 'Categora B', 'Categora C', 'Categora D']
valores_set1 = [15, 30, 45, 60]
valores_set2 = [10, 25, 35, 55]

# Definir el ancho de las barras
ancho_barra = 0.35

# Coordenadas para las barras
x = np.arange(len(categorias))

# Crear el grfico de barras dobles
plt.bar(x - ancho_barra/2, valores_set1, width=ancho_barra, label='Conjunto 1')
plt.bar(x + ancho_barra/2, valores_set2, width=ancho_barra, label='Conjunto 2')

# Etiquetas y ttulo
plt.xlabel('Categoras')
plt.ylabel('Valores')
plt.title('Grfico de Barras Dobles')
plt.xticks(x, categorias)
plt.legend()

# Mostrar el grfico
plt.show()

# Datos para el grfico
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 20]

# Crear el grfico de lneas
plt.plot(x, y, marker='o', linestyle='-')

# Etiquetas y ttulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grfico de Lneas')

# Mostrar el grfico
plt.grid(True)  # Puedes quitar esto si no deseas lneas de cuadrcula
plt.show()

# Definir la funcin lineal
m = 2  # Pendiente
b = 3  # Ordenada al origen

# Generar valores de x
x = np.linspace(0, 10, 100)  # Valores de x de 0 a 10

# Calcular los valores correspondientes de y
y = m * x + b

# Crear el grfico de la funcin
plt.plot(x, y, label=f'y = {m}x + {b}')

# Etiquetas y ttulo
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Grfico de una Funcin Lineal')
plt.legend()

# Mostrar el grfico
plt.grid(True)  # Puedes quitar esto si no deseas lneas de cuadrcula
plt.show()

# Datos para el grfico de pastel
categorias = ['Categora A', 'Categora B', 'Categora C', 'Categora D']
valores = [25, 30, 15, 10]

# Colores para las porciones del pastel
colores = ['#ff9999', '#66b3ff', '#99ff99', '#ffcc99']

# Destacar una porcin (opcional)
explode = (0.1, 0, 0, 0)

# Crear el grfico de pastel
plt.pie(valores, labels=categorias, colors=colores, explode=explode, autopct='%1.1f%%', startangle=90)

# Ttulo
plt.title('Grfico de Pastel')

# Mostrar el grfico
plt.axis('equal')  # Aspecto igual para que el grfico sea un crculo
plt.show()
```
